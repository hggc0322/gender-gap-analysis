```python
df18["SicCodesClean"] = df18.apply(lambda row : str(row["SicCodes"]).replace("\r\n",''), axis=1)
df18["SicCodesSplit"] = df18["SicCodesClean"].str.split(",", n = -1)
df18["SicCodes1"] = df18.apply(lambda row: row["SicCodesSplit"][0], axis=1)
df18["SicCodes2"] = df18.apply(lambda row: '' if len(row["SicCodesSplit"])<2 else row["SicCodesSplit"][1], axis=1)
```

```python
# clean addres
df18["AddressClean"] = df18.apply(lambda row : row["Address"].replace("\r\n",'').replace(" ","+"), axis=1)
df18["inLondon"] = df18.Address.str.contains("London")
df18["AddressSplit"] = df18.Address.str.split(",\r\n", n = -1)
```

```python
# address split
df18["postcode"] = df18.apply(lambda row: row["AddressSplit"][-1], axis=1)
df18["SecondLastAddressString"] = df18.apply(lambda row: row["AddressSplit"][-2], axis=1)
df18["ThirdLastAddressString"] = df18.apply(lambda row: '' if len(row["AddressSplit"])<3 else row["AddressSplit"][-3], axis=1)
df18["FourthLastAddressString"] = df18.apply(lambda row: '' if len(row["AddressSplit"])<4 else row["AddressSplit"][-4], axis=1)
df18["FifthLastAddressString"] = df18.apply(lambda row: '' if len(row["AddressSplit"])<5 else row["AddressSplit"][-5], axis=1)
```

```python
regions = ['City of London', 'Westminster', 'Camden', 'Islington', 'Wandsworth',
       'Hackney', 'Richmond upon Thames', 'Merton', 'Brent', 'Hounslow',
       'Hillingdon', 'Bexley', 'Greenwich', 'Lambeth', 'Havering', 'Lewisham',
       'Southwark', 'Kensington and Chelsea', 'Bromley',
       'Barking and Dagenham', 'Ealing', 'Barnet']
```

```python
df18SicZipGap.columns
```

```python
keys = ['region','top1', 'top2', 'top3', 'bottom1', 'bottom2', 'bottom3']
```

```python
data = []
for region in regions:
    top = list(df18SicZipGap[df18SicZipGap['Unnamed: 1_x']==region].sort_values(by='DiffMedianHourlyPercent')['EmployerName'][:3].values)
    bottom = list(df18SicZipGap[df18SicZipGap['Unnamed: 1_x']==region].sort_values(by='DiffMedianHourlyPercent', ascending=False)['EmployerName'][:3].values)
    row = [region] + top + bottom
    
    data.append(row)
    
```

```python
output = pd.DataFrame(data, columns=keys).merge(gender_gap_data, how='left',
                                        right_on='Unnamed: 1_x',
                                        left_on='region'
                                       )
```

```python
output.columns
```
